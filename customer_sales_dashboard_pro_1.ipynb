```python
import matplotlib.pyplot as plt
import seaborn as sns

chart_title = {"family": "Dejavu Serif"}
plt.figure(figsize=(16,8))


# Chart 1 - Sales by City(Bar chart)
plt.subplot(2,3,1)
plt.grid(axis="y", linestyle="--", alpha=0.5)

bar_colors = ["orange", "blue", "green", "red", "purple"]
citywise_total_sales = df.groupby("City")["Total_Sales"].sum().sort_values(ascending=False)


plt.bar(citywise_total_sales.index, citywise_total_sales.values, color= bar_colors, edgecolor="black", lw =0.6)

plt.xlabel("City")
plt.ylabel("Total Sales")
plt.xticks(rotation=30, size= 8)
plt.yticks(size= 8)
plt.title("Sales by City", fontdict=chart_title)



# Chart 2 - Revenue by Category (Pie)
plt.subplot(2,3,2)
category_sales = df.groupby("Product_Category")["Total_Sales"].sum()

x = category_sales.index
y = category_sales.values

plt.pie(y, startangle=270, autopct="%1.1f%%")
plt.title("Revenue by Product Category")
plt.legend(x, loc="lower left", framealpha=0.5, edgecolor="black", fancybox= True)
plt.axis("equal")


# Chart 3 – Age Distribution (Histogram)
plt.subplot(2,3,3)
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.hist(df["Age"], bins= 5, edgecolor= "black", color="skyblue" )
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution", fontdict=chart_title)



# Chart 4 – Payment Method Distribution(pie)
plt.subplot(2,3,4)

y =df.groupby("Payment_Method").size()
x = y.index
plt.pie(y, labels= x, startangle= 90, autopct="%1.1f%%")
plt.title("Payment Method Distribution", fontdict= chart_title)
plt.legend(x, loc="lower left", framealpha=0.5, edgecolor="black", fancybox= True)
plt.axis("equal")



# Chart 5 – Age vs Total Sales (Scatter Plot)
plt.subplot(2,3,5)
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.scatter(df["Age"], df["Total_Sales"], alpha= 0.7,edgecolor="black", linewidth=0.5)
plt.xlabel("Age")
plt.ylabel("Total Sales")
plt.title("Age vs Total Sales", fontdict= chart_title)



# Chart 6 – Rating by Product Category (Boxplot)
plt.subplot(2,3,6)
plt.grid(axis="y",linestyle="--", alpha=0.5)

sns.boxplot(data= df, x= "Product_Category", y= "Rating")
plt.xlabel("Product Category")
plt.ylabel("Rating")
plt.xticks(rotation=30)
plt.title("Rating by Product Category", fontdict= chart_title)


# Overall Title
plt.suptitle("Customer Sales Analysis Dashboard", fontsize=18, fontweight="bold")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()


```
